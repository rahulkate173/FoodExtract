```python
MODEL_NAME = "google/gemma-3n-E2B-it" # the instruct fine tuned model by GOOGLE
```

```python
## creating and inference with Pipeline API
from transformers import pipeline
import torch
pipe = pipeline(
    "image-text-to-text",
    model = MODEL_NAME,
    device="cuda",
    torch_dtype=torch.bfloat16
)

## model loaded for inference
```

```python
path = "https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/p-blog/candy.JPG"
text = "What animal is on the candy?"
```

```python
prompt = [
    {
        "role":"system",
        "content":"you are and vision model that interupt the image and give detailed expression"
    },
    {
        "role":"user",
        "content":[
            {"type":"image","url":path},
            {"type":"text","text":text}
        ]
    }
]
```

```python
output = pipe(text=prompt, max_new_tokens=200)
print(output[0]["generated_text"][-1]["content"])
```
